```python
import random

def sample_floats(low, high, k):
    """ Return a k-length list of unique random floats
        in the range of low <= x <= high
    """
    result = []
    seen = set()
    for i in range(k):
        x = random.uniform(low, high)
        while x in seen:
            x = random.uniform(low, high)
        seen.add(x)
        result.append(x)
    return result
```

```python
##LR
#Keel
df['LR_Keel_Plate_Length'] = df.apply(
    lambda row: (row.Ship_Rule_Length * 710), axis=1)

df['LR_Keel_Plate_Breadth'] = df.apply(
    lambda row: 1200, axis=1)

df['LR_Keel_Plate_Area'] = df.apply(
    lambda row: (row.LR_Keel_Plate_Length * row.LR_Keel_Plate_Breadth), axis=1)

df['LR_Keel_Plate_Thickness'] = df.apply(
    lambda row: 11, axis=1)


df['LR_Keel_Plate_Quantity'] = df.apply(
    lambda row: 1, axis=1)

df['LR_Keel_Plate_Volume'] = df.apply(
    lambda row: (row.LR_Keel_Plate_Area * row.LR_Keel_Plate_Thickness* row.LR_Keel_Plate_Quantity), axis=1)

df['LR_Keel_Plate_Weight'] = df.apply(
    lambda row: (row.LR_Keel_Plate_Volume *7.85/1000000000), axis=1)

keel = pd.concat([df.iloc[:, 0:3], df.iloc[:,-1]], axis=1)

#Bottom
df['LR_Bottom_Plate_Length'] = df.apply(
    lambda row: (row.Ship_Rule_Length * 710), axis=1)

df['LR_Bottom_Plate_Breadth'] = df.apply(
    lambda row: (1.04/2)*row.Ship_Breadth*1000, axis=1)

df['LR_Bottom_Plate_Area'] = df.apply(
    lambda row: (row.LR_Bottom_Plate_Length * row.LR_Bottom_Plate_Breadth), axis=1)

df['LR_Bottom_Plate_Thickness'] = df.apply(
    lambda row: 9, axis=1)


df['LR_Bottom_Plate_Quantity'] = df.apply(
    lambda row: 2, axis=1)

df['LR_Bottom_Plate_Volume'] = df.apply(
    lambda row: (row.LR_Bottom_Plate_Area * row.LR_Bottom_Plate_Thickness* row.LR_Bottom_Plate_Quantity), axis=1)

df['LR_Bottom_Plate_Weight'] = df.apply(
    lambda row: (row.LR_Bottom_Plate_Volume *7.85/1000000000), axis=1)

bottom = pd.concat([keel, df.iloc[:,-1]], axis=1)

#Inner Bottom

df['LR_Inner_Bottom_Plate_Length'] = df.apply(
    lambda row: (row.Ship_Rule_Length * 660), axis=1)

df['LR_Inner_Bottom_Plate_Breadth'] = df.apply(
    lambda row: (0.82)*row.Ship_Breadth*1000, axis=1)

df['LR_Inner_Bottom_Plate_Area'] = df.apply(
    lambda row: (row.LR_Inner_Bottom_Plate_Length * row.LR_Inner_Bottom_Plate_Breadth), axis=1)

df['LR_Inner_Bottom_Plate_Thickness'] = df.apply(
    lambda row: 9, axis=1)


df['LR_Inner_Bottom_Plate_Quantity'] = df.apply(
    lambda row: 2, axis=1)

df['LR_Inner_Bottom_Plate_Volume'] = df.apply(
    lambda row: (row.LR_Inner_Bottom_Plate_Area * row.LR_Inner_Bottom_Plate_Thickness* row.LR_Inner_Bottom_Plate_Quantity), axis=1)

df['LR_Inner_Bottom_Plate_Weight'] = df.apply(
    lambda row: (row.LR_Inner_Bottom_Plate_Volume *7.85/1000000000), axis=1)

inner_bottom = pd.concat([bottom, df.iloc[:,-1]], axis=1)

#Side Shell

df['LR_Side_Shell_Plate_Length'] = df.apply(
    lambda row: (row.Ship_Rule_Length * 710), axis=1)

df['LR_Side_Shell_Plate_Breadth'] = df.apply(
    lambda row: (0.79)*row.Ship_Draft*1000, axis=1)

df['LR_Side_Shell_Plate_Area'] = df.apply(
    lambda row: (row.LR_Side_Shell_Plate_Length * row.LR_Side_Shell_Plate_Breadth), axis=1)

df['LR_Side_Shell_Plate_Thickness'] = df.apply(
    lambda row: 8, axis=1)


df['LR_Side_Shell_Plate_Quantity'] = df.apply(
    lambda row: 2, axis=1)

df['LR_Side_Shell_Plate_Volume'] = df.apply(
    lambda row: (row.LR_Side_Shell_Plate_Area * row.LR_Side_Shell_Plate_Thickness* row.LR_Side_Shell_Plate_Quantity), axis=1)

df['LR_Side_Shell_Plate_Weight'] = df.apply(
    lambda row: (row.LR_Side_Shell_Plate_Volume *7.85/1000000000), axis=1)

side_shell = pd.concat([inner_bottom, df.iloc[:,-1]], axis=1)

#Bulkhead

df['LR_Bulkhead_Length'] = df.apply(
    lambda row: (row.Ship_Breadth * 1000), axis=1)

df['LR_Bulkhead_Breadth'] = df.apply(
    lambda row: (1.03)*row.Ship_Draft*1000, axis=1)

df['LR_Bulkhead_Area'] = df.apply(
    lambda row: (row.LR_Bulkhead_Length * row.LR_Bulkhead_Breadth), axis=1)

df['LR_Bulkhead_Thickness'] = df.apply(
    lambda row: 7, axis=1)


df['LR_Bulkhead_Quantity'] = df.apply(
    lambda row: 2, axis=1)

df['LR_Bulkhead_Volume'] = df.apply(
    lambda row: (row.LR_Bulkhead_Area * row.LR_Bulkhead_Thickness* row.LR_Bulkhead_Quantity), axis=1)

df['LR_Bulkhead_Weight'] = df.apply(
    lambda row: (row.LR_Bulkhead_Volume *7.85/1000000000), axis=1)

bulkhead = pd.concat([side_shell, df.iloc[:,-1]], axis=1)

#Main Deck Plate

df['LR_Main_Deck_Length'] = df.apply(
    lambda row: str('--'), axis=1)

df['LR_Main_Deck_Breadth'] = df.apply(
    lambda row: str('--'), axis=1)

df['LR_Main_Deck_Area'] = df.apply(
    lambda row: (0.71*row.Ship_Rule_Length * row.Ship_Breadth*0.5*1000*1000), axis=1)

df['LR_Main_Deck_Thickness'] = df.apply(
    lambda row: 8, axis=1)


df['LR_Main_Deck_Quantity'] = df.apply(
    lambda row: 2, axis=1)

df['LR_Main_Deck_Volume'] = df.apply(
    lambda row: (row.LR_Main_Deck_Area * row.LR_Main_Deck_Thickness* row.LR_Main_Deck_Quantity), axis=1)

df['LR_Main_Deck_Weight'] = df.apply(
    lambda row: (row.LR_Main_Deck_Volume *7.85/1000000000), axis=1)

LR = pd.concat([bulkhead, df.iloc[:,-1]], axis=1)
```
